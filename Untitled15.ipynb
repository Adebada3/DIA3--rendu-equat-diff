```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error

#########################################
# Partie 1: Modlisation mathmatique
#########################################

# Dfinition des paramtres initiaux
r = 0.05  # taux de croissance
K = 100000  # capacit maximale (taille du march)
P0 = 1000  # population initiale
dt = 1  # pas de temps (1 jour)
num_steps = 365  # nombre de jours  simuler

def euler_logistic(P0, r, K, dt, num_steps):
    """
    Implmentation de la mthode d'Euler pour l'quation logistique
    dP/dt = r * P * (1 - P/K)
    """
    # Initialisation des tableaux
    time = np.arange(0, num_steps * dt, dt)
    P = np.zeros(num_steps)
    P[0] = P0

    # Application de la mthode d'Euler
    for i in range(1, num_steps):
        dP = r * P[i-1] * (1 - P[i-1]/K)  # Calcul de la drive
        P[i] = P[i-1] + dt * dP  # Mthode d'Euler

    return time, P

# Calcul de la solution avec Euler
time_euler, P_euler = euler_logistic(P0, r, K, dt, num_steps)

# Visualisation de la solution d'Euler
plt.figure(figsize=(10, 6))
plt.plot(time_euler, P_euler, label='Solution Euler')
plt.xlabel('Temps (jours)')
plt.ylabel('Nombre d\'utilisateurs')
plt.title('Croissance des utilisateurs - Mthode d\'Euler')
plt.grid(True)
plt.legend()
plt.show()

#########################################
# Partie 2: Solution analytique
#########################################

def analytical_logistic(P0, r, K, t):
    """
    Solution analytique de l'quation logistique
    P(t) = K / (1 + ((K-P0)/P0) * exp(-r*t))
    """
    return K / (1 + (K-P0)/P0 * np.exp(-r*t))

# Calcul de la solution analytique
P_analytical = analytical_logistic(P0, r, K, time_euler)

# Comparaison des solutions
plt.figure(figsize=(10, 6))
plt.plot(time_euler, P_euler, 'b-', label='Solution Euler')
plt.plot(time_euler, P_analytical, 'r--', label='Solution analytique')
plt.xlabel('Temps (jours)')
plt.ylabel('Nombre d\'utilisateurs')
plt.title('Comparaison des solutions Euler et analytique')
plt.grid(True)
plt.legend()
plt.show()

# Calcul de la MSE entre les deux solutions
mse_solutions = mean_squared_error(P_analytical, P_euler)
print(f"\nMSE entre les solutions Euler et analytique: {mse_solutions:.2f}")

#########################################
# Partie 3: Analyse des donnes relles
#########################################

try:
    # Lecture des donnes
    df = pd.read_csv("Dataset_nombre_utilisateurs.csv")

    # 1. Trac des donnes relles
    plt.figure(figsize=(10, 6))
    plt.plot(df['Jour'], df['Utilisateurs'], 'b-', label='Donnes relles')
    plt.xlabel('Jour')
    plt.ylabel('Nombre d\'utilisateurs')
    plt.title('volution relle du nombre d\'utilisateurs')
    plt.grid(True)
    plt.legend()
    plt.show()

    # 2. Analyse de la saturation
    K_reel = df['Utilisateurs'].max()
    saturation_seuil = 0.95 * K_reel
    moitie_saturation = 0.5 * K_reel

    # Identification des jours de saturation
    jour_saturation = df[df['Utilisateurs'] >= saturation_seuil]['Jour'].iloc[0]
    jour_moitie = df[df['Utilisateurs'] >= moitie_saturation]['Jour'].iloc[0]

    print(f"\nAnalyse de la saturation:")
    print(f"Saturation (95%) atteinte au jour: {jour_saturation}")
    print(f"50% de saturation atteint au jour: {jour_moitie}")

    # 3. Calcul des MSE sur diffrents intervalles
    intervals = [(0, 30), (30, 90), (90, 180), (180, 365)]

    # Ajustement des prdictions  la longueur des donnes relles
    t_real = np.arange(len(df))
    predictions_euler = euler_logistic(P0, r, K, 1, len(df))[1]
    predictions_analytical = analytical_logistic(P0, r, K, t_real)

    print("\nMSE par intervalle:")
    for start, end in intervals:
        mask = (df['Jour'] >= start) & (df['Jour'] <= end)
        mse_euler = mean_squared_error(df[mask]['Utilisateurs'], predictions_euler[mask])
        mse_analytical = mean_squared_error(df[mask]['Utilisateurs'], predictions_analytical[mask])

        print(f"\nIntervalle {start}-{end} jours:")
        print(f"MSE Euler: {mse_euler:.2f}")
        print(f"MSE Analytique: {mse_analytical:.2f}")

    # 4. Trac des trois courbes
    plt.figure(figsize=(12, 6))
    plt.plot(df['Jour'], df['Utilisateurs'], 'b-', label='Donnes relles')
    plt.plot(t_real, predictions_euler, 'r--', label='Modle Euler')
    plt.plot(t_real, predictions_analytical, 'g--', label='Modle Analytique')
    plt.xlabel('Jour')
    plt.ylabel('Nombre d\'utilisateurs')
    plt.title('Comparaison des donnes relles avec les modles')
    plt.grid(True)
    plt.legend()
    plt.show()

    # Modle modifi avec taux de croissance variable
    def r_variable(t, r_base=0.05):
        """Taux de croissance variable avec composante saisonnire"""
        return r_base * (1 + 0.5 * np.sin(2 * np.pi * t / 365))

    def logistic_modified(P0, K, t):
        """Modle logistique modifi avec taux de croissance variable"""
        P = np.zeros_like(t, dtype=float)
        P[0] = P0
        for i in range(1, len(t)):
            r_t = r_variable(t[i-1])
            P[i] = P[i-1] + r_t * P[i-1] * (1 - P[i-1]/K)
        return P

    # Calcul et visualisation du modle modifi
    predictions_modified = logistic_modified(P0, K, t_real)
    mse_modified = mean_squared_error(df['Utilisateurs'], predictions_modified)

    print(f"\nMSE du modle modifi: {mse_modified:.2f}")

    plt.figure(figsize=(12, 6))
    plt.plot(df['Jour'], df['Utilisateurs'], 'b-', label='Donnes relles')
    plt.plot(t_real, predictions_modified, 'm--', label='Modle modifi')
    plt.xlabel('Jour')
    plt.ylabel('Nombre d\'utilisateurs')
    plt.title('Comparaison avec le modle modifi')
    plt.grid(True)
    plt.legend()
    plt.show()

except FileNotFoundError:
    print("Erreur: Le fichier Dataset_nombre_utilisateurs.csv n'a pas t trouv.")
except Exception as e:
    print(f"Une erreur s'est produite: {str(e)}")
```

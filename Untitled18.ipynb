```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from itertools import product
from matplotlib.ticker import MaxNLocator

# Partie 1: Implémentation de l'algorithme d'Euler pour les équations de Lotka-Volterra
def lotka_volterra_model(alpha, beta, delta, gamma, x0, y0, t):
    """
    Résout le système d'équations de Lotka-Volterra en utilisant la méthode d'Euler

    Paramètres:
    alpha (float): taux de croissance des proies
    beta (float): taux de prédation
    delta (float): taux de conversion des proies en prédateurs
    gamma (float): taux de mortalité des prédateurs
    x0 (float): population initiale des proies
    y0 (float): population initiale des prédateurs
    t (array): tableau des temps

    Retourne:
    x (array): population des proies à chaque instant t
    y (array): population des prédateurs à chaque instant t
    """
    dt = t[1] - t[0]  # pas de temps
    x = np.zeros_like(t)
    y = np.zeros_like(t)
    x[0] = x0
    y[0] = y0

    for i in range(1, len(t)):
        # Équations de Lotka-Volterra:
        # dx/dt = αx - βxy
        # dy/dt = δxy - γy
        dx = alpha * x[i-1] - beta * x[i-1] * y[i-1]
        dy = delta * x[i-1] * y[i-1] - gamma * y[i-1]

        # Méthode d'Euler:
        x[i] = x[i-1] + dt * dx
        y[i] = y[i-1] + dt * dy

        # Vérification de validité (éviter les populations négatives)
        if x[i] < 0 or y[i] < 0 or np.isnan(x[i]) or np.isnan(y[i]):
            return np.full_like(t, np.nan), np.full_like(t, np.nan)

    return x, y

def simuler_et_afficher(alpha=1.0, beta=0.1, delta=0.1, gamma=1.5):
    """
    Simule le modèle de Lotka-Volterra avec les paramètres donnés et affiche les résultats
    """
    x0 = 40  # population initiale de lapins
    y0 = 9   # population initiale de renards
    t = np.linspace(0, 30, 300)  # 30 jours, 300 points

    # Simulation
    x, y = lotka_volterra_model(alpha, beta, delta, gamma, x0, y0, t)

    # Affichage
    plt.figure(figsize=(10, 5))
    plt.plot(t, x, label="Lapins", color='green')
    plt.plot(t, y, label="Renards", color='red')
    plt.xlabel("Temps (jours)")
    plt.ylabel("Population")
    plt.title(f"Modèle de Lotka-Volterra : α={alpha}, β={beta}, δ={delta}, γ={gamma}")
    plt.legend()
    plt.grid(True)
    plt.show()

    return x, y

# Partie 2: Optimisation des paramètres du modèle
def charger_donnees(fichier='populations_lapins_renards.csv'):
    """
    Charge les données réelles des populations de lapins et renards
    """
    df = pd.read_csv(fichier)
    return df

def calculer_mse(y_pred, y_true):
    """
    Calcule l'erreur quadratique moyenne (MSE) entre les valeurs prédites et réelles
    """
    return np.mean((y_pred - y_true) ** 2)

def optimiser_parametres(df):
    """
    Optimise les paramètres du modèle en utilisant un grid search

    Paramètres:
    df (DataFrame): données réelles contenant les colonnes 'date', 'lapin', 'renard'

    Retourne:
    best_params (dict): paramètres optimaux (alpha, beta, delta, gamma)
    best_mse (float): MSE minimale obtenue
    """
    # Extraction des données
    x_true = df['lapin'].values
    y_true = df['renard'].values

    # Créer un tableau de temps correspondant aux données
    t = np.linspace(0, len(df)-1, len(df))

    # Valeurs à tester pour chaque paramètre
    param_values = [1/3, 2/3, 1, 4/3]

    # Population initiale
    x0 = x_true[0]
    y0 = y_true[0]

    # Grid search
    results = []

    # Générer toutes les combinaisons de paramètres
    param_combinations = list(product(param_values, repeat=4))
    total_combinations = len(param_combinations)

    print(f"Optimisation en cours... Test de {total_combinations} combinaisons de paramètres.")

    for i, (alpha, beta, delta, gamma) in enumerate(param_combinations):
        if (i+1) % 20 == 0:
            print(f"Progression: {i+1}/{total_combinations} ({(i+1)/total_combinations*100:.1f}%)")

        # Simulation avec ces paramètres
        x_pred, y_pred = lotka_volterra_model(alpha, beta, delta, gamma, x0, y0, t)

        # Si la simulation a échoué (valeurs NaN), attribuer un MSE élevé
        if np.isnan(x_pred).any() or np.isnan(y_pred).any():
            mse_x = mse_y = mse_total = float('inf')
        else:
            # Calcul de l'erreur pour les lapins
            mse_x = calculer_mse(x_pred, x_true)
            # Calcul de l'erreur pour les renards
            mse_y = calculer_mse(y_pred, y_true)
            # MSE totale
            mse_total = (mse_x + mse_y) / 2

        results.append({
            'alpha': alpha,
            'beta': beta,
            'delta': delta,
            'gamma': gamma,
            'mse_lapin': mse_x,
            'mse_renard': mse_y,
            'mse_total': mse_total
        })

    # Convertir en DataFrame pour faciliter l'analyse
    results_df = pd.DataFrame(results)

    # Trouver les meilleurs paramètres
    best_idx = results_df['mse_total'].idxmin()
    best_params = results_df.iloc[best_idx][['alpha', 'beta', 'delta', 'gamma']].to_dict()
    best_mse = results_df.iloc[best_idx]['mse_total']

    print("\nMeilleurs paramètres trouvés:")
    print(f"Alpha: {best_params['alpha']}")
    print(f"Beta: {best_params['beta']}")
    print(f"Delta: {best_params['delta']}")
    print(f"Gamma: {best_params['gamma']}")
    print(f"MSE Lapins: {results_df.iloc[best_idx]['mse_lapin']}")
    print(f"MSE Renards: {results_df.iloc[best_idx]['mse_renard']}")
    print(f"MSE Totale: {best_mse}")

    return best_params, best_mse, results_df

def afficher_resultats_optimaux(df, best_params):
    """
    Affiche les résultats du modèle avec les paramètres optimaux
    et les compare aux données réelles
    """
    # Extraction des données
    x_true = df['lapin'].values
    y_true = df['renard'].values

    # Créer un tableau de temps correspondant aux données
    t = np.linspace(0, len(df)-1, len(df))

    # Population initiale
    x0 = x_true[0]
    y0 = y_true[0]

    # Simulation avec les paramètres optimaux
    x_pred, y_pred = lotka_volterra_model(
        best_params['alpha'],
        best_params['beta'],
        best_params['delta'],
        best_params['gamma'],
        x0, y0, t
    )

    # Affichage
    plt.figure(figsize=(12, 8))

    # Tracé des lapins
    plt.subplot(2, 1, 1)
    plt.plot(t, x_true, 'o', color='green', alpha=0.5, label="Données réelles (lapins)")
    plt.plot(t, x_pred, '-', color='darkgreen', label="Modèle (lapins)")
    plt.xlabel("Temps (jours)")
    plt.ylabel("Population des lapins")
    plt.title("Comparaison du modèle Lotka-Volterra optimisé avec les données réelles")
    plt.legend()
    plt.grid(True)

    # Tracé des renards
    plt.subplot(2, 1, 2)
    plt.plot(t, y_true, 'o', color='red', alpha=0.5, label="Données réelles (renards)")
    plt.plot(t, y_pred, '-', color='darkred', label="Modèle (renards)")
    plt.xlabel("Temps (jours)")
    plt.ylabel("Population des renards")
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Afficher la trajectoire dans l'espace des phases
    plt.figure(figsize=(8, 6))
    plt.plot(x_true, y_true, 'o', color='blue', alpha=0.5, label="Données réelles")
    plt.plot(x_pred, y_pred, '-', color='purple', label="Modèle")
    plt.xlabel("Population des lapins")
    plt.ylabel("Population des renards")
    plt.title("Espace des phases: Renards vs Lapins")
    plt.legend()
    plt.grid(True)
    plt.show()

def afficher_carte_chaleur(results_df):
    """
    Affiche une carte de chaleur pour visualiser l'impact des différents paramètres sur la MSE
    """
    # Pour chaque couple de paramètres, créer une visualisation 2D
    param_pairs = [
        ('alpha', 'beta'),
        ('alpha', 'delta'),
        ('alpha', 'gamma'),
        ('beta', 'delta'),
        ('beta', 'gamma'),
        ('delta', 'gamma')
    ]

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()

    for i, (param1, param2) in enumerate(param_pairs):
        # Pour chaque paire, calculer la MSE moyenne pour toutes les combinaisons des autres paramètres
        pivot = pd.pivot_table(
            results_df,
            values='mse_total',
            index=param1,
            columns=param2,
            aggfunc='mean'
        )

        # Afficher la carte de chaleur
        im = axes[i].imshow(pivot, cmap='viridis_r')

        # Ajouter les étiquettes
        param_values = [1/3, 2/3, 1, 4/3]
        param_labels = ['1/3', '2/3', '1', '4/3']

        axes[i].set_xticks(range(len(param_values)))
        axes[i].set_yticks(range(len(param_values)))
        axes[i].set_xticklabels(param_labels)
        axes[i].set_yticklabels(param_labels)

        # Ajouter les titres
        axes[i].set_xlabel(param2)
        axes[i].set_ylabel(param1)
        axes[i].set_title(f'MSE moyenne selon {param1} et {param2}')

        # Ajouter une barre de couleur
        plt.colorbar(im, ax=axes[i])

    plt.tight_layout()
    plt.show()

def analyser_resultats(results_df):
    """
    Analyse l'impact de chaque paramètre sur la MSE
    """
    params = ['alpha', 'beta', 'delta', 'gamma']

    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    axes = axes.flatten()

    for i, param in enumerate(params):
        # Grouper par valeur du paramètre et calculer la MSE moyenne
        grouped = results_df.groupby(param)['mse_total'].mean().reset_index()

        # Tracer
        axes[i].plot(grouped[param], grouped['mse_total'], 'o-', color='blue')
        axes[i].set_xlabel(param)
        axes[i].set_ylabel('MSE moyenne')
        axes[i].set_title(f'Impact de {param} sur la MSE')
        axes[i].grid(True)

        # Forcer le nombre de ticks sur l'axe x
        axes[i].xaxis.set_major_locator(MaxNLocator(4))

    plt.tight_layout()
    plt.show()

def main():
    # Partie 1: Test du modèle avec des paramètres par défaut
    print("Partie 1: Implémentation de l'algorithme d'Euler")
    print("Simulation avec paramètres par défaut (α=1.0, β=0.1, δ=0.1, γ=1.5)")
    simuler_et_afficher()

    # Partie 2: Optimisation des paramètres
    print("\nPartie 2: Optimisation des paramètres du modèle")
    df = charger_donnees()
    print(f"Données chargées: {len(df)} observations")

    # Afficher les premières lignes des données
    print("Aperçu des données:")
    print(df.head())

    # Effectuer l'optimisation des paramètres
    best_params, best_mse, results_df = optimiser_parametres(df)

    # Afficher les résultats avec les paramètres optimaux
    afficher_resultats_optimaux(df, best_params)

    # Analyser l'impact de chaque paramètre
    analyser_resultats(results_df)

    # Afficher la carte de chaleur
    afficher_carte_chaleur(results_df)

    # Conclusion
    print("\nConclusion:")
    print("Le modèle de Lotka-Volterra a été implémenté avec succès en utilisant l'algorithme d'Euler.")
    print(f"Les paramètres optimaux (α={best_params['alpha']}, β={best_params['beta']}, "
          f"δ={best_params['delta']}, γ={best_params['gamma']}) ont été trouvés en minimisant l'erreur "
          f"quadratique moyenne (MSE={best_mse:.2f}).")

    # Interprétation des paramètres
    print("\nInterprétation des paramètres:")
    print(f"- Alpha ({best_params['alpha']:.2f}): représente le taux de croissance naturel des lapins")
    print(f"- Beta ({best_params['beta']:.2f}): représente le taux de prédation des lapins par les renards")
    print(f"- Delta ({best_params['delta']:.2f}): représente l'efficacité avec laquelle les renards convertissent les lapins capturés en nouveaux renards")
    print(f"- Gamma ({best_params['gamma']:.2f}): représente le taux de mortalité naturel des renards")

if __name__ == "__main__":
    main()
```
